```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model, preprocessing
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from kagglehub import dataset_download
```

```python
le = preprocessing.LabelEncoder()
le
```

# The Data

```python
data = pd.read_csv(f'{dataset_download("mirichoi0218/insurance")}/insurance.csv')
data
```

# Null Values

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
for col in data.columns:
    fig, ax = plt.subplots()
    plt.title(f"{col} histogram")
    sns.histplot(data=data, x=col, ax=ax)
    plt.show()
```

```python
data.smoker = le.fit_transform(data.smoker)
data.sex = le.fit_transform(data.sex)
data.region = le.fit_transform(data.region)
data
```

```python
data.info()
```

# Split the data to train and test data

```python
Y = data.charges
data.drop('charges', axis=1, inplace=True)
X = np.asarray(data)
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

# Perform Data Scaling using Robust Scaler

```python
rs = RobustScaler()
X_Train = rs.fit_transform(X_Train)
X_Test = rs.fit_transform(X_Test)
```

# Linear Regression Model

```python
Model = linear_model.LinearRegression()
Model.fit(X_Train, Y_Train)
```

```python
Model.score(X_Train, Y_Train)
```

```python
Model.score(X_Test, Y_Test)
```

```python
Y_Pred = Model.predict(X_Test)
Y_Test_Pred = pd.DataFrame({"Y_Test": Y_Test, "Y_Pred": Y_Pred})
Y_Test_Pred.head()
```

```python
Y_Test_Pred = Y_Test_Pred.reset_index()
```

```python
plt.figure(figsize=(15, 8))
Y_Test_Pred = Y_Test_Pred.drop(["index"], axis=1)
plt.plot(Y_Test_Pred)
plt.legend(["Actual", "Predicted"])
```

```python
reg_score = r2_score(Y_Test, Y_Pred)
reg_score
```

```python
mse = mean_squared_error(Y_Test, Y_Pred)
print("mse: ", mse)
mape = mean_absolute_percentage_error(Y_Test, Y_Pred)
print("mape: ", mape)
mae = mean_absolute_error(Y_Test, Y_Pred)
print("mae: ", mae)
reg_mse = mean_squared_error(Y_Test, Y_Pred)
reg_rmse = np.sqrt(reg_mse)
print("reg_mse: ", reg_rmse)
```
